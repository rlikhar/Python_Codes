```python
# Importing important Libraries
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Machine learning libraries

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
# Reading the data file
```

```python
df = pd.read_excel('DS-assignment/DS - Assignment Part 1 data set.xlsx')
```

```python
# AS Transaction Date Does have have anny significant Role in data.
# We can exclude it from dataframe
```

```python
df.drop(['Transaction date'], axis=1, inplace=True)
```

```python
# Data Preperation for traing and testing the Model
```

```python
X = df.drop(['House price of unit area'], axis=1)
Y = df['House price of unit area']
```

```python
# Splitting data into training and test data in ratio 8:2
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2, random_state=123)
```

```python
LR = LinearRegression()
rfr = RandomForestRegressor()
GBR = GradientBoostingRegressor()
```

```python
# Training the Linear Regression Model and Checking the results
```

```python
LR.fit(x_train, y_train)
```

```python
y_predict = LR.predict(x_test)
```

```python
mse_LR = mean_squared_error(y_test, y_predict)
mae_LR = mean_absolute_error(y_test, y_predict)
LR_r2 = r2_score(y_test, y_predict)
```

```python
resultsLR = pd.DataFrame({'Mean Squared Error':mse_LR,
                          'Mean Absolute Error':mae_LR,
                          'R Squared Score': LR_r2,
                          }
                           ,index=['Linear Regressor'])

resultsLR
```

```python
# Training the Random Forest Regressor and Checking the result
```

```python
rfr.fit(x_train, y_train)
```

```python
rfr_pred = rfr.predict(x_test)
```

```python
mse_rfr = mean_squared_error(y_test, rfr_pred)
mae_rfr = mean_absolute_error(y_test, rfr_pred)
rfr_r2 = r2_score(y_test, rfr_pred)
```

```python
resultsRFR = pd.DataFrame({'Mean Squared Error':mse_rfr,
                          'Mean Absolute Error':mae_rfr,
                          'R Squared Score': rfr_r2,
                          }
                           ,index=['Random Forest Regressor'])

results2 = pd.concat([resultsRFR,resultsLR])
results2
```

```python
# Training the Gradient Boost Regressor and Checking the result
```

```python
GBR.fit(x_train, y_train)
```

```python
gbr_pred = GBR.predict(x_test)
```

```python
mse_gbr = mean_squared_error(y_test, gbr_pred)
mae_gbr = mean_absolute_error(y_test, gbr_pred)
gbr_r2 = r2_score(y_test, gbr_pred)
```

```python
# Data Frame for comparision of Different Models result
```

```python
resultsGBR = pd.DataFrame({'Mean Squared Error':mse_gbr,
                          'Mean Absolute Error':mae_gbr,
                          'R Squared Score': gbr_r2,
                          }
                           ,index=['Gradient Boost Regressor'])

results3 = pd.concat([resultsGBR,results2])
results3
```

```python
# AS we find that the MSE is minimum for GradientBoost regressor so
# Hyper Parameter tuning using GridSearchCV
```

```python
search_grid={'n_estimators':[500,1000,2000],
             'learning_rate':[.001,0.01,.1],
             'max_depth':[1,2,4],
             'subsample':[.5,.75,1],
             'random_state':[1]}
```

```python

gridCV = GridSearchCV(estimator=GBR,
                      param_grid=search_grid,
                      scoring='neg_mean_squared_error',
                      n_jobs=1,
                      cv=5)
```

```python
gridCV.fit(x_train, y_train)
```

```python
print(gridCV.best_params_)
```

```python
print(gridCV.best_score_)
```

```python
# Training the best model obtained using GridSearch CV and getting the result
```

```python
final_GBR = GradientBoostingRegressor(learning_rate=0.01, max_depth=4, n_estimators=500, subsample=0.75, random_state=1)
```

```python
final_GBR.fit(x_train, y_train)
```

```python
pred = final_GBR.predict(x_test)
```

```python
mse = mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test,pred)
r2 = r2_score(y_test, pred)
```

```python
result = pd.DataFrame({'Mean Squared Error':mse,
                          'Mean Absolute Error':mae,
                          'R Squared Score':r2,
                          }
                           ,index=['Final Gradient Boost Regressor'])

results4 = pd.concat([result,results3])
results4
```
